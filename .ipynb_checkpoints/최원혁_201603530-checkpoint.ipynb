```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from datetime import datetime

%matplotlib inline
```

```python
onlineRetail = pd.read_csv("OnlineRetail-data.csv")
onlineRetail.head(5)
```

# Line Chart

```python
onlineRetail.Description.value_counts()[:10].plot(figsize=(30,15))
```

**많이 팔린 상위 10개의 물품명을 Line Chart로 나타냅니다.**

# Bar Chart

```python
def toDtObject(value):
    return datetime.strptime(value,'%m/%d/%Y %H:%M')
onlineRetail['InvoiceDate']=onlineRetail['InvoiceDate'].apply(toDtObject)
def getMonth(value):
    return int(value.strftime('%m'))
onlineRetail['month']=onlineRetail['InvoiceDate'].apply(getMonth)
onlineRetail.month.value_counts().plot(figsize=(15,6), kind="bar")
plt.show()
```

**판매가 많은 월을 순서대로 Bar Chart로 나타냅니다. 11월에 블랙프라이데이, 추수감사절, 12월에 크리스마스, 연말 등의 이벤트가 있는 것을 미루어보아 판매량이 많은 것을 추측할 수 있습니다.**

# Scatter Plot

```python
def getHour(value):
    return int(value.strftime('%H'))
onlineRetail['hour']=onlineRetail['InvoiceDate'].apply(getHour)
onlineRetail.plot.scatter(x='hour', y='UnitPrice', figsize=(15,6))
```

**시간대 별 판매된 품목 단가를 Scatter Plot으로 나타냅니다. 12~15시 정도 오후 시간대일 수록 단가가 높은 품목의 판매가 이뤄지고 있음을 알 수 있습니다.**

# Box Plot

```python
onlineRetail.Country.value_counts()[-20:].plot(figsize=(15,6), kind='box')
plt.show()
```

**판매 빈도 하위 20개 국가를 Box Plot으로 나타냅니다. 하위 20개 국가에서 평균 140번 정도의 판매가 이뤄지고 있음을 알 수 있습니다.**
