해당 URL을 오픈하여 내용을 bs 라는 객체에 저장 

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=018&aid=0004535411'
html = urlopen(url)  

bs = BeautifulSoup(html, "html.parser") 

# print(bs) # 해당 웹페이지 전체를 html 그대로 출력 
print(bs.head.title)
```

### bs 객체의 head 내에서 meta라는 태그를 모두 검색하여 content를 출력 

```python
for meta in bs.head.find_all('meta'):
    print(meta.get('content'))
```

## bs의 body 내에서 div 태그를 찾아 그중 id가 articleBodyContents 인 것을 찾아 텍스트를 출력

```python
print(bs.body.find('div', {'id':'articleBodyContents'}).get_text())
```

### 페이지의 모든 링크를 가져와 리스트에 저장 

```python
links = []

idx = 0
for link in bs.find_all('a'):
    l_text = link.text.strip()
    l_href = link.get('href')
    idx += 1
    dict_link = {'addr':l_href, 'text':l_text}
    links.append(dict_link)
    print(str(idx), ': ', l_text, l_href)
```

### 페이지에서 news.naver.com의 링크만 가져와 리스트에 저장 

```python
links = []

idx = 0
for link in bs.find_all('a'):
    l_text = link.text.strip()
    l_href = link.get('href')
    if l_href.startswith('https://news.naver.com/main/read.nhn'):
        idx += 1
        dict_link = {'addr':l_href, 'text':l_text}
        links.append(dict_link)
        print(str(idx), ': ', l_text, l_href)
```

```python
import pandas as pd

df = pd.DataFrame(links)
```

### 위의 링크를 돌면서 내용을 데이터프레임에 저장

```python
news_list = []

for link in links:
    url = link['addr']
    title = link['text']
    html = urlopen(url)  
    bs = BeautifulSoup(html, "html.parser") 
    content = bs.body.find('div', {'id':'articleBodyContents'}).get_text()
    content = content.replace('\n\n\n\n\n// flash 오류를 우회하기 위한 함수 추가\nfunction _flash_removeCallback() {}','')
    content = content.replace('\n',' ')
    content = content.replace('\t',' ')
    dict_news = {'addr':url, 'text':title, 'content':content} 
    news_list.append(dict_news)

news_contents = pd.DataFrame(news_list)
```
