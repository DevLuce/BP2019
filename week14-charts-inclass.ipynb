## formatting line type and color
- color: 
  - matplotlib의 line color 관련 내용 (https://matplotlib.org/tutorials/colors/colors.html) 
  - simply use one of {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}
- marker type:
  - matplotlib의 marker 관련 내용 (https://matplotlib.org/3.1.1/api/markers_api.html#module-matplotlib.markers)

```python
x = np.arange(0,10) 
y = x ^ 2 
#Labeling the Axes and Title
plt.title("Graph Drawing") 
plt.xlabel("Time") 
plt.ylabel("Distance") 

# Formatting the line colors
plt.plot(x,y,marker='>',markerfacecolor='b')

# # Formatting the line type  
# plt.plot(x,y,'>') 

# save in pdf formats
plt.savefig('timevsdist.pdf', format='pdf')
```

### Annotation

```python
plt.figure()

x = np.arange(0,10) 
y = x ^ 2 
z = x ^ 3
t = x ^ 4 
# Labeling the Axes and Title
plt.title("Graph Drawing") 
plt.xlabel("Time") 
plt.ylabel("Distance") 
plt.plot(x,y)

# #Annotate
plt.annotate(xy=[2,1], s='Second Entry') 
plt.annotate(xy=[4,6], s='Third Entry') 
plt.annotate(xy=[8,10], s='Tenth Entry') 
```

### Legend

```python
plt.figure()

x = np.arange(0,10) 
y = x ^ 2 # XOR exclusive OR operation  (10)
z = x ^ 3 # (11)
t = x ^ 4 # (1000)
# Labeling the Axes and Title
plt.title("Graph Drawing") 
plt.xlabel("Time") 
plt.ylabel("Distance") 
plt.plot(x,y)

#Annotate
plt.annotate(xy=[2,1], s='Second Entry') 
plt.annotate(xy=[4,6], s='Third Entry') 
# Adding Legends
plt.plot(x,z)
plt.plot(x,t)

legend = ['Race1', 'Race2','Race3']

plt.legend(legend, loc=4) 
```

```python
plt.figure()

x = np.arange(0,10) 
y = x ^ 2 
z = x ^ 3
t = x ^ 4 
# Labeling the Axes and Title
plt.title("Graph Drawing") 
plt.xlabel("Time") 
plt.ylabel("Distance") 
plt.plot(x,y)

#Annotate
plt.annotate(xy=[2,1], s='Second Entry') 
plt.annotate(xy=[4,6], s='Third Entry') 
# Adding Legends
plt.plot(x,z)
plt.plot(x,t)
plt.legend(['Race1', 'Race2','Race3'], loc=4) 

#Style the background
plt.style.use('seaborn-bright')
plt.plot(x,z)
```

## Drawing a 3D Plot


```python
chart = plt.figure()
chart3d = chart.add_subplot(111, projection='3d')

# Create some test data.
X, Y, Z = axes3d.get_test_data(0.08)

# Plot a wireframe.
chart3d.plot_wireframe(X, Y, Z, color='r',rstride=15, cstride=10)

plt.show()
 
```
